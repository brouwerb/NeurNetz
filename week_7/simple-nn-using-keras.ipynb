```python
import numpy as np
import pylab
#from keras.models import Sequential
#from keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
N=5000
x1 = 10*np.random.rand(N)-5*np.ones(N)  # random numbers [-5,5]
x2 = 10*np.random.rand(N)-5*np.ones(N)

ytrain=np.zeros(N)
for i in range(N):
    ytrain[i]=1+np.sign(x1[i]*x2[i])/2
```

```python
net=Sequential()
net.add(Dense(4,input_shape=(2,),activation='relu'))
net.add(Dense(4,activation='relu'))
#net.add(Dense(10,activation='relu'))
net.add(Dense(1,activation='linear'))
net.compile(loss='mean_squared_error', optimizer='adam')
x=np.transpose((x1,x2))
```

```python
ypredict=np.ravel(net.predict_on_batch(x))
pylab.scatter(x1,x2,c=ypredict)
pylab.show()
```

```python
net.fit(x, ytrain, batch_size=100, epochs=50, validation_split=0.1)
```

```python
ypredict=np.ravel(net.predict_on_batch(x))
pylab.scatter(x1,x2,c=ypredict)
pylab.show()
```
