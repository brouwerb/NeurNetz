```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np

xres = 1000
tr = 2000
c = 20
tres = 0.9/(xres*c)
wav = np.zeros([tr, xres])
x0 = np.linspace(0, 1, xres)
XX, YY = np.meshgrid(x0, np.linspace(0, tres*tr, tr))


wav[0] = (5*np.sin(4*np.pi*x0))
for i in range(1, len(wav)):
    for j in range(1, len(wav[0])-1):
        wav[i][j] = 2*wav[i-1][j] - wav[i-2][j] + c**2*(tres**2 * xres**2)*(wav[i-1][j+1] - 2*wav[i-1][j] + wav[i-1][j-1])


fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(XX, YY, wav)
display(fig)
```
